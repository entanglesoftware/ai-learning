```python
!pip install langchain langchain-openai

import os
from google.colab import drive
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

#  Mount Google Drive
drive.mount('/content/drive')

#  Define API Key File Path
api_key_path = "/content/drive/My Drive/OpenAi-key/OpenAi-key.txt"


if os.path.exists(api_key_path):
    with open(api_key_path, "r", encoding="utf-8-sig") as file:
        os.environ["OPENAI_API_KEY"] = file.read().strip()


# Initialize OpenAI Chat Model
chat = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.7)

# Create Prompt Template
blog_template = PromptTemplate(
    input_variables=["topic"],
    template="""
Write a **detailed and professional blog** on "{topic}".

**Guidelines:**
- Introduction
- Headings and subheadings
- Provide an engaging introduction.
- Use structured paragraphs with real-world examples.
- Discuss future predictions and ethical concerns.
- Summarize key takeaways in the conclusion.

Write in **clear and fluent English** without repeating sentences.

---

### {topic}
"""
)

# Define LangChain LLM Chain
blog_chain = LLMChain(llm=chat, prompt=blog_template)

# Generate Blog Content
topic = "The Future of AI in Healthcare"
blog_content = blog_chain.invoke({"topic": topic})

# Print Generated Blog
print("\nGenerated Blog:\n")
print(blog_content["text"])
```
